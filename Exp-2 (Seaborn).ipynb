```python
b=sns.load_dataset("tips")
b.head(7)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=b)
```

```python
sns.relplot(data=b, x="total_bill", y="tip", hue="day")
```

Assigning a col variable creates a faceted figure with multiple subplots arranged across the columns of the grid:

```python
sns.relplot(data=b, x="total_bill", y="tip", hue="day", col="time")
```

Different variables can be assigned to facet on both the columns and rows:

```python
sns.relplot(data=b, x="total_bill", y="tip", hue="day", col="time", row="sex")
```

Assigning multiple semantic variables can show multi-dimensional relationships, but be mindful to avoid making an overly-complicated plot.

```python
sns.relplot(data=b, x="total_bill", y="tip", col="time", hue="time", style="sex", palette=["b", "r"])
```

```python
c= sns.load_dataset("fmri")
c.head()
```

Using kind="line" offers the same flexibility for semantic mappings as kind="scatter", but lineplot() transforms the data more before plotting.

```python
sns.relplot(data=c, x="timepoint", y="signal", col="region", hue="event", style="event", kind="line")
```

The size and shape of the figure is parametrized by the height and aspect ratio of each individual facet:

```python
sns.relplot(data=c, x="timepoint", y="signal", col="region", hue="event", style="event", kind="line", height=4, aspect=.7)
```

# Bar Plots :

```python
titanic=sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
```

```python
sns.catplot(x="deck", kind="count", palette="ch:.25", data=titanic)
```

```python
sns.catplot(y="deck", hue="class", kind="count",palette="pastel", edgecolor=".6",data=titanic)
```

# Point plots

```python
sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic)
```

```python
sns.catplot(x="class", y="survived", hue="sex", palette={"male":"g","female":"m"}, markers=["^","o"], linestyles=["-","--"], kind="point", data=titanic)
```

```python
g=sns.catplot(x="fare", y="survived", row="class", kind='box', orient='h', height=1.5, aspect=4, data=titanic.query("fare>0"))
```

# Histograms and Kernel density estimation:

```python
data= np.random.multivariate_normal([0,0],[[5,2],[2,2]], size=2000)
```

```python
data=pd.DataFrame(data, columns=['x','y'])
```

HISTOGRAM

KDEPLOT :

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

Histograms and KDE can be combined using displot :

```python
sns.distplot(data['x'])
sns.distplot(data['y'])
```
